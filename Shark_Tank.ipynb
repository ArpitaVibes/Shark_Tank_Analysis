```python
columns_to_keep = ['Season Number','Episode Number','Startup Name','Industry','Business Description','Pitchers City','Pitchers State','Started in','Number of Presenters','Male Presenters','Female Presenters','Pitchers Average Age','Original Ask Amount','Original Offered Equity',
                   'Valuation Requested','Received Offer','Accepted Offer','Total Deal Amount','Total Deal Equity','Total Deal Debt','Debt Interest','Deal Valuation','Namita Investment Amount','Vineeta Investment Amount','Anupam Investment Amount','Aman Investment Amount','Peyush Investment Amount',
                   'Ritesh Investment Amount','Amit Investment Amount','Guest Investment Amount','Royalty Percentage','Bootstrapped','Has Patents']
                   
```

```python
df_filtered = df[columns_to_keep].copy()
```

```python
df_filtered.isnull().sum().sort_values(ascending=False)
```

```python
df_filtered['Has Patents'] = df_filtered['Has Patents'].fillna(False)
df_filtered['Bootstrapped'] = df_filtered['Bootstrapped'].fillna(False)
df_filtered['Female Presenters'] = df_filtered['Female Presenters'].fillna(0)
df_filtered['Male Presenters'] = df_filtered['Male Presenters'].fillna(0)
df_filtered['Pitchers City'] = df_filtered['Pitchers City'].fillna("Unknown")
df_filtered['Pitchers State'] = df_filtered['Pitchers State'].fillna("Unknown")
```

```python
df_filtered = df_filtered.dropna(subset=['Started in'])
```

```python
investor_cols = ['Namita Investment Amount','Vineeta Investment Amount','Anupam Investment Amount',
                 'Aman Investment Amount','Peyush Investment Amount','Ritesh Investment Amount',
                 'Amit Investment Amount','Guest Investment Amount']
```

```python
df_filtered[investor_cols] = df_filtered[investor_cols].fillna(0)
```

```python
#deal success by founder gender composition
```

```python
def classify_team(row):
    if row['Male Presenters'] > 0 and row['Female Presenters'] > 0:
        return 'Mixed'
    elif row['Female Presenters'] > 0:
        return 'Female Only'
    elif row['Male Presenters'] > 0:
        return 'Male Only'
    else:
        return 'Unknown'
```

```python
df_filtered['Team Gender'] = df_filtered.apply(classify_team, axis=1)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_filtered, x='Team Gender', y='Accepted Offer', hue='Team Gender')
plt.title('Deal Success Rate by Team Gender')
plt.ylabel('Proportion of Accepted Offers')
plt.xlabel('Team Gender Composition')
#plt.ylim(0,1)
plt.show()
```

```python
#Deal success by startup age group
```

```python
bins = [0,2,5,10,20]
labels = ['0-2 yrs','3-5 yrs','6-10 yrs','10+ yrs']
df_filtered['Startup Age Group'] = pd.cut(df_filtered['Startup Age'], bins=bins, labels=labels)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_filtered, x='Startup Age Group', y= 'Accepted Offer', color='seagreen')
plt.title('Deal Success by Startup Age Group')
plt.ylabel('Proportion of Accepted Offers')
plt.xlabel('Startup Age Group')
plt.ylim(0,1)
plt.show()
```
